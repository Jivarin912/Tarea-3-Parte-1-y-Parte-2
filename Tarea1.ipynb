1. Load an image containing **at least three different objects from different classes** using `cv2.imread`.

```python
# Write your code here...
import cv2
from matplotlib import pyplot as plt
image_bgr = cv2.imread(r'H:\MaestriaUIDE\clase_vision_computador_3\data\images.jpg')
print('Image shape:', image_bgr.shape)
# Convert BGR to RGB for displaying with matplotlib
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

# Display the image
plt.imshow(image_rgb)
plt.show()
```

2. Draw **different colored bounding boxes** around each object (at least one of the bounding boxes should be yellow). Use a **line thickness** of 5 for both bounding boxes.

```python
# Write your code here...

# Perro (AMARILLO)
cv2.rectangle(image_bgr, (10, 30), (120, 230), (0, 255, 255), 5)

# Niño (AZUL)
cv2.rectangle(image_bgr, (100, 20), (200, 230), (255, 0, 0), 5)

# Niña (VERDE)
cv2.rectangle(image_bgr, (170, 40), (280, 230), (0, 255, 0), 5)

# Convertir BGR → RGB
img_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)




```

3. Add **text labels** above each bounding box (e.g., “Cat”, “Ball”, etc.) using `cv2.putText`.

```python
# Write your code here...

#Perro (AMARILLO)
cv2.putText(image_bgr, "Perro", (10, 70), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 255), 2)

# Niño (AZUL)
cv2.putText(image_bgr, "Boy", (100, 50), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 0, 0), 2)

# Niña (VERDE)
cv2.putText(image_bgr, "Girl", (190, 70), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)

# Convertir BGR → RGB
img_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

```

4. Display the **resulting image**.
